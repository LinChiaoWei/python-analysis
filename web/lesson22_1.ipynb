```python
from google import genai
from dotenv import load_dotenv
import os
load_dotenv()


client = genai.Client(api_key=os.environ['Gemini_API_KEY'])

response = client.models.generate_content(
    model="gemini-2.5-flash", contents="天空為什麼是藍的"
)
print(response.text)
```
