```python
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.font_manager import fontManager

fontManager.addfont("ChineseFont.ttf")
mpl.rc('font', family="ChineseFont")

student1 = {'國文':78, '英文':80, '數學':92, '歷史':75, '探索':85}
student2 = {'國文':90, '英文':75, '數學':98, '歷史':65, '探索':75}
subjects = list(student1.keys()) #dict的實體方法:keys.values.items
stu1_values = list(student1.values()) #轉為list
stu2_values = list(student2.values()) 
subjects, stu1_values, stu2_values#顯示為tuple
```

```python
#標準寫法
figure = plt.figure(figsize=(8,5),dpi=240,facecolor='#cccccc')
axes = figure.add_subplot(1,1,1)
axes.set_title("學生分數")
axes.set_xlabel("科目")
axes.set_ylabel("分數")
axes.plot(subjects,stu1_values,marker='o',linestyle='-.',label="學生1",alpha=0.8)#alpha為透明度
axes.plot(subjects,stu2_values,marker='o',linestyle='-.',label="學生2",alpha=0.8)
axes.hlines(85,xmin=-1,xmax=6,color='red',linestyles='--',alpha=0.3)
x_locations = axes.get_xticks()
y_locations = axes.get_yticks()
#print(axes.get_xticklabels())
#axes.set_xticklabels(['chiness','english','math','history','discover'])
axes.set_xlim(xmin=min(x_locations)-1,xmax=max(x_locations)+1)
axes.set_ylim(ymin=min(y_locations)-10,ymax=max(y_locations)+10)

for x in x_locations:
    axes.annotate(str(stu1_values[x]),(x,stu1_values[x]+1))
    axes.annotate(str(stu2_values[x]),(x,stu2_values[x]+1))
#axes.annotate("註解",(0,80))
axes.legend()
axes.grid(axis='y')
plt.show()
plt.close()
```
